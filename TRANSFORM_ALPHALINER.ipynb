ubah tipe data, hitung ulang presentase, dan simpan data yang sudah di *transform*

```python
df.columns = df.columns.str.strip()

numeric_cols = [
    "Total TEU", "Total Ships",
    "Owned TEU", "Owned Ships",
    "Chartered TEU", "Chartered Ships",
    "Orderbook TEU", "Orderbook Ships"
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace(",", "", regex=False)
            .replace({"": "0", "nan": "0"})
            .astype(float)
        )

percent_cols = ["% Chartered", "% Existing"]

for col in percent_cols:
    if col in df.columns:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace("%", "", regex=False)
            .replace({"": "0", "nan": "0"})
            .astype(float)
        )

        df = df.fillna(0)

df['% Chartered'] = (df['Chartered TEU'] / df['Total TEU'] * 100).round(2)
df['% Existing'] = (df['Orderbook TEU'] / df['Total TEU'] * 100).round(2)
display(df)

df.to_csv('alphalinertop100_tabel2_transform.csv', index=False)

```
